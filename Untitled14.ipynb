```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('nature2.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')
plt.show()
```

```python
def stretch_contrast(img):
    low = np.min(img)
    high = np.max(img)
    result = ((img - low) * (255 / (high - low))).astype(np.uint8)
    return result

stretched = stretch_contrast(image)
```

```python
equalized = cv2.equalizeHist(image)
```

```python
fig, ax = plt.subplots(3, 2, figsize=(12, 12))


ax[0, 0].imshow(image, cmap='gray')
ax[0, 0].set_title('Original Image')
ax[0, 0].axis('off')
ax[0, 1].hist(image.ravel(), bins=256, color='black')
ax[0, 1].set_title('Histogram of Original')

ax[1, 0].imshow(stretched, cmap='gray')
ax[1, 0].set_title('Contrast Stretched')
ax[1, 0].axis('off')
ax[1, 1].hist(stretched.ravel(), bins=256, color='black')
ax[1, 1].set_title('Histogram of Stretched')


ax[2, 0].imshow(equalized, cmap='gray')
ax[2, 0].set_title('Histogram Equalized')
ax[2, 0].axis('off')
ax[2, 1].hist(equalized.ravel(), bins=256, color='black')
ax[2, 1].set_title('Histogram of Equalized')

plt.tight_layout()
plt.show()
```
